```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('miss').getOrCreate()
```

```python
df = spark.read.csv('ContainsNull.csv',header=True,inferSchema=True)
```

```python
df.show()
```

```python
df.na.drop().show()
```

```python
df.na.drop(thresh=2).show() # parametr thresh=2, jeśli są conajmniej dwie wartości nie null to nie usuwa
```

```python
df.na.drop(how='any').show() # 'any' is deafult, deleting all rows where is null value
```

```python
df.na.drop(how='all').show()
```

```python
df.na.drop(subset=['Sales']).show() # usuwa null po kolumnie
```

```python
# Filling missing Values

df.na.fill('No Name', subset=['Name']).show()
```

```python
from pyspark.sql.functions import mean
mean_val = df.select(mean(df['Sales'])).collect()
mean_sales = mean_val[0][0]
df.na.fill(mean_sales,['Sales']).show()
```
